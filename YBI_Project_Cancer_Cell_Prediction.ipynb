# **Project Title: Cancer Cell Prediction**


Dataset Source: from YBI Foundation (GitHub)


**Objective: To predict if a cell is cancerous based on given parameters**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/Cancer.csv')
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.shape
```

```python
df['diagnosis'].value_counts()
```

```python
df.groupby('diagnosis').mean()
```

```python
 # Define y (dependent variable)
 y = df['diagnosis']
```

```python
y.shape
```

```python
y
```

```python
# x = independent variable
x = df[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
x.shape
```

```python
x
```

```python
# Standardization of x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ss = MinMaxScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

```python
# Train Test Split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.5, stratify=y, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#  Model Train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
#Model Prediction
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
# Probability of Predicted class
```

```python
lr.predict_proba(x_test)
```

```python
#Model Evaluations
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
